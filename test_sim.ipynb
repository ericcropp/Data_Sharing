```python
for key in I.input:
    if key not in I_orig.input:
        print(f"Key '{key}' is missing in I_orig.input")
    else:
        if I.input[key] != I_orig.input[key] and key !='fieldmaps':
            if isinstance(I.input[key], dict) and isinstance(I_orig.input[key], dict):
                for subkey in I.input[key]:
                    if subkey not in I_orig.input[key]:
                        print(f"  Subkey '{subkey}' is missing in I_orig.input[{key}]")
                    else:
                        if I.input[key][subkey] != I_orig.input[key][subkey]:
                            print(f"  Difference found in key:subkey '{key}':'{subkey}':")
                            print(f"  I.input[{key}][{subkey}]: {I.input[key][subkey]}")
                            print(f"  I_orig.input[{key}][{subkey}]: {I_orig.input[key][subkey]}")
            print(f"Difference found in key '{key}':")
            print(f"I.input[{key}]: {I.input[key]}")
            print(f"I_orig.input[{key}]: {I_orig.input[key]}")


# Compare I.input['lattice'] and I_orig.input['lattice'] by 'name' key, order does not matter
lattice_I_dict = {elem.get('name', f'idx_{i}'): elem for i, elem in enumerate(I.input['lattice'])}
lattice_I_orig_dict = {elem.get('name', f'idx_{i}'): elem for i, elem in enumerate(I_orig.input['lattice'])}

all_names = set(lattice_I_dict.keys()).union(lattice_I_orig_dict.keys())

for name in all_names:
    if name not in lattice_I_dict:
        print(f"Element '{name}' is missing in I.input['lattice']")
    elif name not in lattice_I_orig_dict:
        print(f"Element '{name}' is missing in I_orig.input['lattice']")
    else:
        if lattice_I_dict[name] != lattice_I_orig_dict[name]:
            for subkey in lattice_I_dict[name]:
                if subkey not in lattice_I_orig_dict[name]:
                    print(f"    Subkey '{subkey}' is missing in I_orig.input['lattice'][{name}]")
                elif lattice_I_dict[name][subkey] != lattice_I_orig_dict[name][subkey]:
                    print(f"    Difference found in subkey '{subkey}':")
                    print(f"      I.input['lattice'][{name}][{subkey}]: {lattice_I_dict[name][subkey]}")
                    print(f"      I_orig.input['lattice'][{name}][{subkey}]: {lattice_I_orig_dict[name][subkey]}")
            for subkey in lattice_I_orig_dict[name]:
                if subkey not in lattice_I_dict[name]:
                    print(f"    Subkey '{subkey}' is missing in I.input['lattice'][{name}]")
            print(f"Difference found in lattice element '{name}':")
            print(f"  I.input['lattice'][{name}]: {lattice_I_dict[name]}")
            print(f"  I_orig.input['lattice'][{name}]: {lattice_I_orig_dict[name]}")
```
